```python
def f(x, y):
    return  np. cos(X*Y*100)  * np.cos(x)
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 90)
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
fig = plt.figure()
ax = plt.axes(projection ='3d')

ax.contour3D(X, Y, Z)
```

```python
def f(x, y):
    return  np. cos(X*Y*100)  * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5, 90)

X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X, Y, Z, colors='gray');
```

```python
plt.contourf(X, Y, Z, 30,cmap='gray');
plt.colorbar();
```

```python
plt.contourf(X, Y, Z, 30, cmap='gray')
plt.colorbar();
```

```python
plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='gray')
plt.colorbar()
#plt.axis(n='image');
```

```python
contours = plt.contour(X, Y, Z, 3, colors='black')
plt.clabel(contours, inline=True, fontsize=8)

plt.imshow(Z, extent=[0, 5, 0, 5], origin='lower',
           cmap='gray', alpha=0.5)
plt.colorbar();
```
